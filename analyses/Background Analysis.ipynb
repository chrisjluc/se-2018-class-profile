# Background Analysis


## Setup

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
plt.style.use('ggplot')

# Show matplotlib plots in this notebook
%matplotlib inline
# Setting plot parameters
from pylab import rcParams
params = {
    'figure.figsize': (8, 8),
    'legend.fontsize': 15
}
rcParams.update(params)
```

```python
df = pd.read_csv('../data/results.csv')
```

```python
def bar_chart(col):
    df.groupby(col).size().sort_index(ascending=False).plot(kind='bar')
    plt.show()

def pie_chart(col):
    df.groupby(col).size().sort_index(ascending=False).plot(kind='pie', subplots=True)
    plt.show()
    
def bar_chart_vs_salary(col):
    df[[col, 'coop_salary_avg']].groupby(col).mean().plot(kind='bar')
    plt.show()
```

## Analysis

```python
bar_chart('gender')
bar_chart('location')
bar_chart('is_international')
bar_chart('ethnicity')
bar_chart('parents_edu')
bar_chart('parents_technical')
bar_chart('family_income')
bar_chart('admission_avg')
```

```python
bar_chart_vs_salary('gender')
bar_chart_vs_salary('location')
bar_chart_vs_salary('is_international')
bar_chart_vs_salary('ethnicity')
bar_chart_vs_salary('parents_edu')
bar_chart_vs_salary('parents_technical')
bar_chart_vs_salary('family_income')
bar_chart_vs_salary('admission_avg')
```
