# Background Analysis


## Setup

```python
from IPython.display import display

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
plt.style.use('ggplot')

# Show matplotlib plots in this notebook
%matplotlib inline
# Setting plot parameters
from pylab import rcParams
params = {
    'figure.figsize': (8, 8),
    'legend.fontsize': 15
}
rcParams.update(params)
```

```python
df = pd.read_csv('../data/results.csv')
```

```python
parents_edu_order = ['Did not complete high school', 'High school', 'Some postsecondary', 
                          'Postsecondary diploma (e.g. college)', 'Bachelor\'s', 
                          'Master\'s/Professional (MD, MBA, etc)', 'Doctorate']
family_income_order = ['0-50k', '50k-100k', '100k-150k', '150k-200k', '200k-250k', '>250k']
rounded_admission_avg_order = [i for i in range(70, 101)]
```

```python
def bar_chart(col, order=None):
    if order:
        results = df.groupby(col).size().reindex(order)
    else:
        results = df.groupby(col).size().sort_values(ascending=False)
    display(results.to_frame())
    results.plot(kind='bar')
    plt.show()
       
def bar_chart_x_vs_y(x, y, order=None):
    if order:
        results = df[[x, y]].groupby(x).mean().reindex(order)
    else:
        results = df[[x, y]].groupby(x).mean().sort_values(y, ascending=False)
    display(results)
    results.plot(kind='bar')
    plt.show()
    
def bar_chart_percentile_x_vs_y(col, y, num_percentiles=10):
    percentiles = np.linspace(0, 1, num=num_percentiles + 1)
    
    # drop first percentile because we'll check if less than than that value
    percentile_values = list(df[col].quantile(percentiles))[1:] 
    percentiles = [int(100 * p) for p in percentiles[1:]]
    
    # copy df because we're going to add a temp column
    d = df[[col, y]].copy()
    
    def get_percentile(x):
        for percentile, percentile_value in zip(percentiles, percentile_values):
            if x <= percentile_value:
                return percentile
            
    def get_percentile_value(x):
        for percentile, percentile_value in zip(percentiles, percentile_values):
            if x <= percentile_value:
                return percentile_value
    
    # create new percentile column
    percentile_col = col + '_percentile'
    percentile_value_col = col + '_percentile_value'
    d[percentile_col] = d[col].apply(get_percentile)
    d[percentile_value_col] = d[col].apply(get_percentile_value)
    
    # generate results
    results = d[[percentile_col, percentile_value_col, y]].groupby(percentile_col).mean().reindex(percentiles)
    display(results)
    results.plot(kind='bar', x=percentile_value_col, y=y)
    plt.show()
```

## Analysis

```python
bar_chart('gender')
bar_chart('location')
bar_chart('is_international')
bar_chart('ethnicity')
bar_chart('parents_edu', parents_edu_order)
bar_chart('parents_technical')

bar_chart('family_income', family_income_order)
print 'Number of people who weren\'t sure or prefered not to disclose family income: ' + \
        str(df.loc[df['family_income'] == 'Not sure / prefer not to disclose'].shape[0])
    
bar_chart('rounded_admission_avg', rounded_admission_avg_order)
```

```python
bar_chart_x_vs_y('gender', 'coop_salary_avg')
bar_chart_x_vs_y('location', 'coop_salary_avg')
bar_chart_x_vs_y('is_international', 'coop_salary_avg')
bar_chart_x_vs_y('ethnicity', 'coop_salary_avg')
bar_chart_x_vs_y('parents_edu', 'coop_salary_avg', parents_edu_order)
bar_chart_x_vs_y('parents_technical', 'coop_salary_avg')
bar_chart_x_vs_y('family_income', 'coop_salary_avg', family_income_order)
bar_chart_x_vs_y('rounded_admission_avg', 'coop_salary_avg', rounded_admission_avg_order)
bar_chart_percentile_x_vs_y('admission_avg', 'coop_salary_avg')
```

```python
bar_chart_x_vs_y('gender', 'culm_grade_avg')
bar_chart_x_vs_y('location', 'culm_grade_avg')
bar_chart_x_vs_y('is_international', 'culm_grade_avg')
bar_chart_x_vs_y('ethnicity', 'culm_grade_avg')
bar_chart_x_vs_y('parents_edu', 'culm_grade_avg', parents_edu_order)
bar_chart_x_vs_y('parents_technical', 'culm_grade_avg')
bar_chart_x_vs_y('family_income', 'culm_grade_avg', family_income_order)
bar_chart_x_vs_y('rounded_admission_avg', 'culm_grade_avg', rounded_admission_avg_order)
bar_chart_percentile_x_vs_y('admission_avg', 'culm_grade_avg')
```
